# **Utilizando yfinance para ações brasileiras**

A biblioteca possui basicamente três módulos:

* yf.Ticker
* yf.download
* yf.pandas_datareader

Quase todos os métodos estão no módulo yf.Tickers

O módulo **download** serve para baixar rapidamente os dados históricos de vários tickers de uma só vez.

# **Extração de cotação multiticker**

```python
yf.download("GOLL4.SA", start = '2023-01-01', end = '2023-10-01')
```

```python
yf.download(["WEGE3.SA", "GOLL4.SA", "ABEV3.SA", "CRFB3.SA", "NTCO3.SA", "VIIA3.SA"], start = '2023-01-01', end = '2023-10-01')
```

```python
yf.download(["WEGE3.SA", "GOLL4.SA", "ABEV3.SA", "CRFB3.SA", "NTCO3.SA", "VIIA3.SA"], start = '2023-01-01', end = '2023-10-01', group_by = 'column')
```

```python
yf.download(["WEGE3.SA", "GOLL4.SA", "ABEV3.SA", "CRFB3.SA", "NTCO3.SA", "VIIA3.SA"], start = '2023-01-01', end = '2023-10-01', group_by = 'tickers')
```

```python
Percentual = (Valor_Atual-Valor_Min)*100/Valor_Min
Percentual
```
